```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
```

```python
from keras.models import load_model
model = load_model('best_models/fine_tuned')
```

```python
from keras.preprocessing import image

new_img = image.load_img('test/yes_tumor (5).jpg', target_size=(224, 224))
img = image.img_to_array(new_img)
img = np.expand_dims(img, axis=0)
img = img/255
plt.imshow(img[0])
plt.axis('off')
pred = model.predict(img)
print(pred)
if pred[0]>0.966:
    print("Having Tumor")
else:
    print("Not Having Tumor")
```

```python
new_img = image.load_img('test/no_tumor (5).jpg', target_size=(224, 224))
img = image.img_to_array(new_img)
img = np.expand_dims(img, axis=0)
img = img/255
plt.imshow(img[0])
plt.axis('off')
pred = model.predict(img)
print(pred)
if pred[0]>0.966:
    print("Having Tumor")
else:
    print("Not Having Tumor")
```
